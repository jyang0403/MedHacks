```python
books=pd.read_csv('books.csv')
movies=pd.read_csv('movies.csv',encoding='latin-1')
```

```python
wiki_wiki = wikipediaapi.Wikipedia('en')
def find_synopsis(string):
  page_py = wiki_wiki.page(string)
  return page_py.sections[0].text

  
```

```python
movies['Synopsis']=movies['Movie Name']
books['Synopsis']=books['Book Name']
```

```python
! pip install transformers
```

```python
from transformers import pipeline
sentiment_analysis = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")
```

```python
def label(string):
  lst=[]
  for i in (1,len(string)//512):
    lst+=[(sentiment_analysis(string[512*(i-1):512*i])[0]['label'])]
  lst+=[sentiment_analysis(string[512*(i):])[0]['label']]
  return (mode(lst))
```

```python
movies['Synopsis Sentiment']=movies['Synopsis'].apply(label)
```

```python
books['Synopsis Sentiment']=books['Synopsis'].apply(label)
```

```python
movies.to_csv('moviesfinal.csv')
books.to_csv('booksfinal.csv')
```
